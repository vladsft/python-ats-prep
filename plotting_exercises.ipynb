**Task**: Create a simple line plot of a mathematical function.

- Plot the function $y = x^2$ for $x$ values ranging from -10 to 10.
- Label the x-axis as "x" and the y-axis as "y".
- Add a title to the plot: "Plot of $y = x^2$".

```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-10.0, 10.0, 200)
plt.figure(figsize=(5, 3))
plt.plot(xs, [i*i for i in xs])
plt.xlabel('x\'s')
plt.ylabel('their square')
plt.show()
```

**Task**: Create a plot with multiple lines and customize the appearance.

- Plot the functions $y = x^2$ and $y = x^3$ on the same graph for $x$ values ranging from -10 to 10.
- Use different colors and line styles for each function.
- Add a legend to distinguish between the two functions.
- Add grid lines to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-10.0, 10.0, 200)
plt.figure(figsize=(5, 3))
plt.plot(xs, [i*i for i in xs], color="purple", label="square")
plt.plot(xs, [i*i*i for i in xs], color="green", label="cube")
plt.xlabel('x\'s')
plt.ylabel('their functions')
plt.legend()
plt.grid(True)
plt.show()
```

**Task**: Create a figure with multiple subplots and add annotations.

- Create a figure with two subplots arranged in a 1x2 grid.
- In the first subplot, plot the sine function $y = \sin(x)$ for $x$ values ranging from 0 to $$ 2\pi $$.
- In the second subplot, plot the cosine function $y = \cos(x)$ for the same range of $x$ values.
- Add annotations to indicate the maximum and minimum points on each plot.
- Label the axes and add titles to each subplot.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

xs = np.arange(0, math.pi, math.pi/12)
plt.figure(figsize=(10, 10))
plt.plot(xs, [math.sin(i) for i in xs], color="purple", label="sin")
plt.plot(xs, [math.cos(i) for i in xs], color="green", label="cos")
plt.xlabel('x\'s')
plt.ylabel('their functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# GPT's version
import numpy as np
import matplotlib.pyplot as plt

# Define x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot sine function
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.set_title('Sine Function')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Plot cosine function
ax2.plot(x, np.cos(x), label='cos(x)', color='orange')
ax2.set_title('Cosine Function')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

# Define custom x-ticks and labels
ticks = [0, np.pi/6, np.pi/3, np.pi/2, 2*np.pi/3, 5*np.pi/6, np.pi, 7*np.pi/6, 4*np.pi/3, 3*np.pi/2, 5*np.pi/3, 11*np.pi/6, 2*np.pi]
tick_labels = ['0', r'$\pi/6$', r'$\pi/3$', r'$\pi/2$', r'$2\pi/3$', r'$5\pi/6$', r'$\pi$', r'$7\pi/6$', r'$4\pi/3$', r'$3\pi/2$', r'$5\pi/3$', r'$11\pi/6$', r'$2\pi$']

# Apply custom ticks and labels to both subplots
ax1.set_xticks(ticks)
ax1.set_xticklabels(tick_labels)
ax2.set_xticks(ticks)
ax2.set_xticklabels(tick_labels)

# Add grid and legend
ax1.grid(True)
ax2.grid(True)
ax1.legend()
ax2.legend()

# Show plot
plt.tight_layout()
plt.show()
```
